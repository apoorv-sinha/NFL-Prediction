Multilayer Perceptron

```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Activation
import matplotlib.pyplot as plt
```

```python
ann = tf.keras.models.Sequential()
ann.add(tf.keras.layers.Dense(units=64, activation='relu'))
ann.add(tf.keras.layers.Dense(units=32, activation='softmax'))
ann.add(tf.keras.layers.Dense(units=8, activation='softmax'))
ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
ann.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import EarlyStopping
early_stop = EarlyStopping(monitor='loss', mode='max', verbose=1, patience=35)
```

```python
ann.fit(X_train, y_train, batch_size = 16, epochs = 500, verbose = 1, callbacks = [early_stop])
```

```python
y_pred_binary = ann.predict(X_train)
y_pred_binary = y_pred_binary.flatten()
len(y_pred_binary)
```
